Independent and dependent features

```python
X=df.drop('FWI',axis=1)
y=df['FWI']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train.shape,X_test.shape
```

```python
#feature selection based on corelation
```

```python
X_train.corr()
```

```python
#check for multicollinearity independent features must not be that much corelated
```

```python
plt.figure(figsize=(12,10))
corr=X_train.corr()
sns.heatmap(corr,annot=True)
```

morethan 90 percent values which are corelated we have to remove that because independent features must not be that much corelated

```python
def correlation(dataset, threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold:
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
corr_features=correlation(X_train, 0.85)
```

```python
#drop features when corelation is more than 0.85
```

```python
X_train.drop(corr_features,axis=1,inplace=True)
X_test.drop(corr_features,axis=1,inplace=True)
```

```python
X_train.shape,X_test.shape
```

FEATURE SCALING OR STANDARDIZATION

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar=StandardScaler()
```

```python
X_train_SCALED=scalar.fit_transform(X_train)
```

```python
X_test_SCALED=scalar.transform(X_test)
```

```python
X_train_SCALED
```

#box plot to understand the effect of standard scalar

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
sns.boxplot(data=X_train)
plt.subplot(1,2,2)
sns.boxplot(data=X_train_SCALED)
plt.title('x_train after scaling')
```

```python
plt.subplots(figsize=(15,5))

plt.subplot(1,2,1)
sns.boxplot(data=X_train)
plt.subplot(1,2,2)
sns.boxplot(data=X_train_SCALED)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
le=LinearRegression()
le.fit(X_train_SCALED,y_train)
y_pred=le.predict(X_test_SCALED)

r2=r2_score(y_pred,y_test)
mse=mean_squared_error(y_pred,y_test)
print(r2)
print(mse)
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score,mean_squared_error
ridge=Ridge()
ridge.fit(X_train_SCALED,y_train)
y_pred=ridge.predict(X_test_SCALED)

r2=r2_score(y_pred,y_test)
mse=mean_squared_error(y_pred,y_test)
print(r2)
print(mse)
```

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import r2_score,mean_squared_error
lasso=Lasso()
lasso.fit(X_train_SCALED,y_train)
y_pred=lasso.predict(X_test_SCALED)

r2=r2_score(y_pred,y_test)
mse=mean_squared_error(y_pred,y_test)
print(r2)
print(mse)
```

```python
from sklearn.linear_model import ElasticNet
from sklearn.metrics import r2_score,mean_squared_error
elasticnet=ElasticNet()
elasticnet.fit(X_train_SCALED,y_train)
y_pred=elasticnet.predict(X_test_SCALED)

r2=r2_score(y_pred,y_test)
mse=mean_squared_error(y_pred,y_test)
print(r2)
print(mse)
```
